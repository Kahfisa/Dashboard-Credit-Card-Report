## Gathering Data

Data Overview

1. client_num: Nomor unik identifikasi pelanggan (ID unik untuk setiap nasabah).
2. customer_age: Usia pelanggan.
3. gender: Jenis kelamin pelanggan.
4. dependent_count: Jumlah tanggungan pelanggan.
5. education_level: Tingkat pendidikan pelanggan.
6. marital_status: Status pernikahan pelanggan.
7. state_cd: Kode provinsi atau wilayah tempat pelanggan tinggal.
8. zipcode: Kode pos tempat tinggal pelanggan.
9. car_owner: Menunjukkan apakah pelanggan memiliki mobil.
10. house_owner: Menunjukkan apakah pelanggan memiliki rumah.
11. personal_loan: Menunjukkan apakah pelanggan memiliki pinjaman pribadi.
12. contact: Jenis kontak utama pelanggan.
13. customer_job: Jenis pekerjaan atau profesi pelanggan.
14. income: Pendapatan tahunan pelanggan.
15. cust_satisfaction_score: Skor kepuasan pelanggan terhadap layanan bank atau kartu.
16. card_category: Jenis kartu kredit yang dimiliki pelanggan.
17. annual_fees: Biaya tahunan yang dibebankan untuk kartu kredit.
18. activation_30_days: Indikator apakah kartu diaktifkan dalam 30 hari setelah diterbitkan.
19. customer_acq_cost: Biaya akuisisi pelanggan (biaya marketing/sales untuk mendapatkan pelanggan tersebut).
20. week_start_date: Tanggal awal minggu yang menjadi periode pengamatan transaksi atau data.
21. week_num: Nomor minggu dalam tahun.
22. qtr: Kuartal tahun (1-4).
23. current_year: Tahun data dicatat.
24. credit_limit: Batas maksimum kredit yang diberikan kepada pelanggan.
25. total_revolving_bal: Saldo bergulir (jumlah tagihan yang belum dibayar penuh dan diteruskan ke bulan berikutnya).
26. total_trans_amt: Total nilai transaksi yang dilakukan pelanggan dalam periode tertentu.
27. total_trans_vol: Jumlah transaksi (frekuensi penggunaan kartu) dalam periode tertentu.
28. avg_utilization_ratio: Rasio pemanfaatan kredit rata-rata.
29. use_chip: pakah transaksi menggunakan chip kartu.
30. exp_type: Tipe kartu.
31. interested_earned: Bunga atau insentif yang diperoleh.
32. deliquent_acc: Jumlah akun bermasalah atau keterlambatan pembayaran.

## Cleaning Data

```python
columns = ["client_num", "zipcode"]

for col in columns:
    df[col] = df[col].astype("object")
```

```python
columns = ["week_start_date"]

for col in columns:
    df[col] = pd.to_datetime(df[col], format = "%d-%m-%Y")
```

```python
columns = ["income", "annual_fees", "customer_acq_cost", "total_revolving_bal", "total_trans_amt", "total_trans_vol"]

for col in columns:
    df[col] = df[col].astype(float)
```
